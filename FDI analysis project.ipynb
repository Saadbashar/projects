```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
fdi_data = pd.read_csv('FDI data.csv')

# Display the first few rows
fdi_data.head()
```

```python
# Check for missing values
print(fdi_data.isnull().sum())

# Summary of the dataset
fdi_data.describe()
```

```python
# Transpose the data for trend analysis
trend_data = fdi_data.set_index('Sector').T
trend_data.index = trend_data.index.str[:7]  # Extract years

# Plot the total FDI inflows over the years
total_fdi = trend_data.sum(axis=1)
plt.figure(figsize=(10, 6))
sns.lineplot(x=total_fdi.index, y=total_fdi.values, marker='o')
plt.title('Total FDI Inflows Over the Years')
plt.xlabel('Year')
plt.ylabel('FDI Inflows (in million USD)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Calculate total FDI by sector
sector_fdi = fdi_data.set_index('Sector').sum(axis=1).sort_values(ascending=False)

# Plot top 10 sectors attracting FDI
plt.figure(figsize=(12, 6))
sector_fdi.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Sectors by Total FDI Inflows')
plt.xlabel('Sector')
plt.ylabel('Total FDI Inflows (in million USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot trends for the top 5 sectors
top_sectors = sector_fdi.head(5).index
plt.figure(figsize=(12, 6))
for sector in top_sectors:
    sns.lineplot(x=trend_data.index, y=trend_data[sector], label=sector)

plt.title('FDI Trends for Top 5 Sectors')
plt.xlabel('Year')
plt.ylabel('FDI Inflows (in million USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare data for forecasting
years = np.arange(len(total_fdi)).reshape(-1, 1)
fdi_values = total_fdi.values

# Train a linear regression model
model = LinearRegression()
model.fit(years, fdi_values)

# Predict for the next 5 years
future_years = np.arange(len(total_fdi), len(total_fdi) + 5).reshape(-1, 1)
future_fdi = model.predict(future_years)

# Plot actual and predicted FDI
plt.figure(figsize=(10, 6))
plt.plot(total_fdi.index, total_fdi.values, label='Actual', marker='o')
plt.plot(
    [f'{2021 + i}' for i in range(1, 6)],
    future_fdi,
    label='Forecast',
    marker='x',
    linestyle='--',
)
plt.title('FDI Forecast')
plt.xlabel('Year')
plt.ylabel('FDI Inflows (in million USD)')
plt.legend()
plt.grid()
plt.show()
```
